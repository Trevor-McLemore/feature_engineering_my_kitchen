```python
def get_proxies():
    url = 'https://free-proxy-list.net/'
    response = requests.get(url)
    parser = fromstring(response.text)
    proxies = set()
    for i in parser.xpath('//tbody/tr')[:10]:
        if i.xpath('.//td[7][contains(text(),"yes")]'):
            proxy = ":".join([i.xpath('.//td[1]/text()')[0], i.xpath('.//td[2]/text()')[0]])
            proxies.add(proxy)
    return proxies
```

```python
## Finds all the urls, filters them for only product pages, and returns them as a list (redundancies remain)

def filtered_url(soup):
    div = soup.find_all('div', {'class': 'list-elements'})
    url_list = []
    for data in soup.find_all('div', class_='list-elements'):
        for a in data.find_all('a'):
            if "review" in a.get('href'):
                continue
            if "?deals" in a.get('href'):
                continue
            if "?quick_ship" in a.get('href'):
                continue
            if "wishlist" in a.get('href'):
                continue
            if "#save-tag" in a.get('href'):
                continue
            if "#add&IndexID" in a.get('href'):
                continue
            if "redirect" in a.get('href'):
                continue
            if "pre-page" in a.get('href'):
                continue
            if "pricerequest" in a.get('href'):
                continue
            if "#/" in a.get('href'):
                continue
            else:
                url_list.append(a.get('href').replace('https', 'http'))
    return list(url_list)
```

```python
def remove_redundancy(lst):
    return list(set(lst))
```

```python
## checks membereship and creates a dictionary with the data
def add_data_to_dict(feature_list):
    
    d = {}
    
    feature_dict = feature_list[0]
    if 'sku' in feature_dict:
        d['sku'] = feature_dict['sku']
    if 'name' in feature_dict:
        d['item'] = feature_dict['name']
    if 'brand' in feature_dict:
        if 'name' in feature_dict['brand']:
            d['brand'] = feature_dict['brand']['name']
    if 'offers'in feature_dict:
        if 'category'in feature_dict['offers'][0]:
            d['category'] = feature_dict['offers'][0]['category']
    if 'color' in feature_dict:
        d['color'] = feature_dict['color']
    if 'weight' in feature_dict:
        d['weight'] = feature_dict['weight']
    if 'aggregateRating' in feature_dict:
        if 'ratingValue' in feature_dict['aggregateRating']:
            d['rating'] = feature_dict['aggregateRating']['ratingValue']
    if 'aggregateRating' in feature_dict:
        if 'reviewCount' in feature_dict['aggregateRating']:
            d['times_rated'] = feature_dict['aggregateRating']['reviewCount']
    if 'offers'in feature_dict:
        if 'price'in feature_dict['offers'][0]:
            d['price'] = feature_dict['offers'][0]['price']
    
    return d
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
standard_X_train = scaler.transform(X_train)
standard_X_test = scaler.transform(X_test)
```

```python
from sklearn.model_selection import cross_val_score

lasso = Lasso()
cross_val_MSE = cross_val_score(lasso, X_train, y_train, cv = 5, scoring = 'neg_mean_squared_error')
cross_val_RMSE = np.sqrt(-cross_val_MSE)
```

```python
model = Lasso()
model.fit(standard_X_train, y_train)
ypred = model.predict(standard_X_test)
RMSE = mean_squared_error(y_test, ypred, squared = False)
RMSE
```
